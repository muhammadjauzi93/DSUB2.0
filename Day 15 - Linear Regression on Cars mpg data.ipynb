```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/muhammadjauzi93/DSUB2.0/main/auto-mpg.csv")
```

```python
# drop discrete data
drop = ['car name','origin','cylinders']
linear_df = df.drop(drop,axis=1)
sns.pairplot(linear_df)
```

```python
sns.regplot(data=linear_df,x='weight',y='mpg')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#predict mpg using weight
features = ['weight']

train_data = linear_df[features]
train_target = linear_df['mpg']

#split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(train_data,train_target,test_size=0.3, random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#predict car mpg of car with 1000 pounds
model.predict([[1000]])
```
