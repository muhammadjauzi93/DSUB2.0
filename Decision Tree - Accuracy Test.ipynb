```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/theleadio/datascience_demo/master/framingham.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
# Eliminate rows with NA values
nona_df = df.dropna()
nona_df.describe()
```

```python
corr = nona_df.corr()
sns.set(rc={'figure.figsize':(20,20)})
sns.heatmap(corr, annot=True);
```

```python
#Column not used for decision tree
drop = ['education','currentSmoker','TenYearCHD']
```

```python
#Drop column not used for decision tree
train_for_diabetes=nona_df.drop(drop,axis=1)

#Drop target variable
train_for_diabetes=train_for_diabetes.drop('diabetes',axis=1)

#Target variable to predict
train_label = nona_df['diabetes']
```

```python
train_for_diabetes.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=3)
model.fit(train_for_diabetes,train_label);
```

```python
import graphviz
columns = list(train_for_diabetes.columns)
dot_data = tree.export_graphviz(model,out_file=None, feature_names=columns,class_names=['No','Yes'],filled=True, rounded=True)
```

```python
graph = graphviz.Source(dot_data)
graph
```

```python
from sklearn.model_selection import train_test_split

#split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(train_for_diabetes,train_label,test_size=0.3, random_state=1)
```

```python
#accuracy test
model_acc = tree.DecisionTreeClassifier(max_depth=3)
model_acc.fit(X_train,y_train);
```

```python
y_pred = model_acc.predict(X_test)
```

```python
from sklearn import metrics
print(f"Accuracy: {metrics.accuracy_score(y_test,y_pred)}")
```
