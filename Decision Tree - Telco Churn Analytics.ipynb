```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/theleadio/datascience_demo/master/telco_customer_churn_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert TotalCharges column to float64 and replace blanks with NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df.info()
```

```python
df['gender'].unique()
```

```python
#replace binary data values with 1 and 0
binary_columns = ['Partner','Dependents','PaperlessBilling','Churn','PhoneService','gender']

for x in binary_columns:
    df[x] = df[x].map({'No':0,'Yes':1,'Female':0,'Male':1})
```

```python
df.info()
```

```python
df['MultipleLines'] = df['MultipleLines'].map({'No':0,'Yes':1,'No phone service':0})
```

```python
df.info()
```

```python
binary_columns2 = ['DeviceProtection','TechSupport','StreamingMovies','StreamingTV','OnlineBackup','OnlineSecurity']
for x in binary_columns2:
    df[x] = df[x].map({'No':0,'Yes':1,'No internet service':0})
```

```python
df.info()
```

```python
df = pd.get_dummies(data=df,columns=['InternetService'])
```

```python
df.head()
```

```python
df.info()
```

```python
# Eliminate rows with NA values
nona_df = df.dropna()
nona_df.info()
```

```python
corr = nona_df.corr()
sns.set(rc={'figure.figsize':(20,20)})
sns.heatmap(corr, annot=True,fmt='.2f');
```

```python
nona_df.describe()
```

```python
#subplots
from matplotlib import pyplot as plt
```

```python
#subplots for binary data
main, subplots = plt.subplots(2,3, figsize=(16,15))
sns.countplot(data=nona_df,x='gender',hue='Churn', ax=subplots[0,0])
sns.countplot(data=nona_df,x='SeniorCitizen',hue='Churn', ax=subplots[0,1])
sns.countplot(data=nona_df,x='InternetService_Fiber optic',hue='Churn', ax=subplots[0,2])
sns.countplot(data=nona_df,x='Partner',hue='Churn', ax=subplots[1,0])
sns.countplot(data=nona_df,x='Dependents',hue='Churn', ax=subplots[1,1])
sns.countplot(data=nona_df,x='PhoneService',hue='Churn', ax=subplots[1,2])
```

```python
#subplots for continuous data
main, subplots = plt.subplots(1,3, figsize=(16,10))
sns.boxplot(data=nona_df, x='Churn',y='tenure',ax=subplots[0])
sns.boxplot(data=nona_df, x='Churn',y='MonthlyCharges',ax=subplots[1])
sns.boxplot(data=nona_df, x='Churn',y='TotalCharges',ax=subplots[2])
```

```python
#remove outliers row to improve accuracy
#nona_df_noOutliers = nona_df.drop(nona_df[(nona_df.TotalCharges >= 6000) | (nona_df.tenure > 70)].index)
nona_df_noOutliers = nona_df
nona_df_noOutliers.describe()
```

```python
#Column not used for decision tree
drop = ['customerID','Contract','PaymentMethod','Churn']
```

```python
#Drop column not used for decision tree
train_data=nona_df_noOutliers.drop(drop,axis=1)

#Target variable to predict
train_label = nona_df_noOutliers['Churn']
```

```python
train_data.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=3)
model.fit(train_data,train_label);
```

```python
import graphviz
columns = list(train_data.columns)
dot_data = tree.export_graphviz(model,out_file=None, feature_names=columns,class_names=['No','Yes'],filled=True, rounded=True)
```

```python
graph = graphviz.Source(dot_data)
graph
```

```python
from sklearn.model_selection import train_test_split

#split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(train_data,train_label,test_size=0.3, random_state=1)
```

```python
#accuracy test
model_acc = tree.DecisionTreeClassifier(max_depth=3)
model_acc.fit(X_train,y_train);
```

```python
y_pred = model_acc.predict(X_test)
```

```python
from sklearn import metrics
print(f"Accuracy: {round(metrics.accuracy_score(y_test,y_pred)*100,2)} %")
```
